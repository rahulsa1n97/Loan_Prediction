# Train_Test split

```python
train = df.drop(['Loan_ID','Loan_Status'],axis=1)
```

```python
target = df.Loan_Status
```

# 1st model Logistic_regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(train,target)
```

```python
pred = log_reg.predict(train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred,target)*100
```

# 2nd model Decision_Tree_Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(train,target)
```

```python
pred1 = clf.predict(train)
pred1
```

```python
accuracy_score(pred1,target)*100
```

# 3rd model Random_Forest_classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf1 = RandomForestClassifier(max_depth=15, random_state=0)
```

```python
clf1.fit(train,target)
```

```python
pred2 = clf1.predict(train)
```

```python
pred2
```

```python
accuracy_score(pred2,target)*100
```
